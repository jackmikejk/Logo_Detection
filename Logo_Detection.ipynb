# 2. Remove Dodgy Images

```python
import cv2
import imghdr
```

```python
data_directory = 'ddata'
```

```python
image_exsts = ['jpeg', 'jpg', 'bmp', 'png']
```

```python
for img_class in os.listdir(data_directory):
    for image in os.listdir(os.path.join(data_directory, img_class)):
        img_path = os.path.join(data_directory, img_class, image)
        try:
            img = cv2.imread(img_path)
            type_of_image = imghdr.what(img_path)
            if type_of_image not in image_exsts:
                print('Image not in the existing list{}'.format(img_path))
                os.remove(img_path)
        except Exception as e:
            print('Issue with image {}: {}'.format(image_path, e))
    
```
